```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
road = cv2.imread('../DATA/road_image.jpg')
```

```python
roadcpy=np.copy(road)
```

```python
road.shape
```

```python
marker_img=np.zeros(road.shape[:2],dtype=np.int32)
```

```python
segments=np.zeros(road.shape,dtype=np.uint8)
```

```python
from matplotlib import cm
cm.tab10(0)
```

```python
def create_rgb(i):
    return tuple(np.array(cm.tab10(i)[:3])*255)
```

```python
colors=[]
for i in range(10):
    colors.append(create_rgb(i))
```

```python
colors
```

```python
###
current_marker=1
marks_updated=False
n_markers=10
###

###
```

```python
def mouse_callback(event,x,y,flags,param):
    global marks_updated
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(marker_img,(x,y),10,(current_marker),-1)
        cv2.circle(roadcpy,(x,y),10,colors[current_marker],-1)
        marks_updated=True
```

```python
    cv2.namedWindow('Road Image')
    cv2.setMouseCallback('Road Image',mouse_callback)
    
    while True:
        cv2.imshow('Watershed Segments',segments)
        cv2.imshow('Road Image',roadcpy)
        
        ## CLOSE ALL
        k=cv2.waitKey(1)
        if k ==27:
            break
        
        ## Clear colors with C
        elif k==ord('c'):
            roadcpy=road.copy()
            marker_img=np.zeros(road.shape[:2],dtype=np.int32)
            segments=np.zeros(road.shape,dtype=np.uint8)
        elif k>0 and chr(k).isdigit():
            current_marker=int(chr(k))
        
        if marks_updated:
            marker_img_cpy=marker_img.copy()
            cv2.watershed(road,marker_img_cpy)
            segments=np.zeros(road.shape,dtype=np.uint8)
            for color_ind in range(n_markers):
                segments[marker_img_cpy==(color_ind)]=colors[color_ind]
            marks_updated=False
        
cv2.destroyAllWindows()
```
