```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def display (img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
coins=cv2.imread('../DATA/pennies.jpg')
```

```python
display(coins)
```

```python
# MEDIAN BLUR
# GRAY
# BINARY THRESHOLD
# FIND CONTOURS
```

```python
sep_blur=cv2.medianBlur(coins,25)
```

```python
display(sep_blur)
```

```python
gray_sep=cv2.cvtColor(sep_blur,cv2.COLOR_BGR2GRAY)
```

```python
ret,sep_thresh=cv2.threshold(gray_sep,160,255,cv2.THRESH_BINARY_INV)
display(sep_thresh)
```

```python
image, contours, hierarchy = cv2.findContours(sep_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    if(hierarchy[0][i][3])==-1:
        cv2.drawContours(coins,contours,i,(255,0,0),10)
```

```python
## WATER SHED ALGORITHM -------------------------------------
coins=cv2.imread('../DATA/pennies.jpg')
```

```python
coins=cv2.medianBlur(coins,45)
```

```python
display(coins)
```

```python
gray=cv2.cvtColor(coins,cv2.COLOR_BGR2GRAY)
```

```python
#USAR OTSU
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
display(thresh)
```

```python
#NOISE REMOVAL
kernel=np.ones((3,3),np.uint8)
```

```python
opening=cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations=2)
```

```python
display(opening)
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)
```

```python
dist = cv2.distanceTransform(opening,cv2.DIST_L2,5)
```

```python
display(dist)
```

```python
ret,sure_fore= cv2.threshold(dist,0.7*dist.max(),255,0)
```

```python
display(sure_fore)
```

```python
sure_fore=np.uint8(sure_fore)
```

```python
unknown=cv2.subtract(sure_bg,sure_fore)
```
