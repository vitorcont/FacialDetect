```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
flat_chess=cv2.imread('../DATA/flat_chessboard.png')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
gray_flat=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
found,corners=cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
plt.imshow(flat_chess)
```

```python
real_chess=cv2.imread('../DATA/real_chessboard.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
gray_real=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)

found,corners=cv2.findChessboardCorners(real_chess,(8,3))

cv2.drawChessboardCorners(real_chess,(8,3),corners,found)
plt.imshow(real_chess)
```
