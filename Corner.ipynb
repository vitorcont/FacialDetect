```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
flat_chess=cv2.imread('../DATA/flat_chessboard.png')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat,cmap='gray')
```

```python
real_chess=cv2.imread('../DATA/real_chessboard.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
gray=np.float32(gray_flat)
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst=cv2.dilate(dst,None)
flat_chess[dst>0.01*dst.max()] = [255,0,0] #VERMEI
```

```python
plt.imshow(flat_chess)
```

```python
gray=np.float32(gray_real)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst=cv2.dilate(dst,None)
real_chess[dst>0.01*dst.max()] = [255,0,0] #VERMEI
```

```python
plt.imshow(real_chess)
```

```python
#aula 2------------------------------------
flat_chess=cv2.imread('../DATA/flat_chessboard.png')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
gray_flat=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)

real_chess=cv2.imread('../DATA/real_chessboard.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
gray_real=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners=cv2.goodFeaturesToTrack(gray_flat,10,0.01,10)
corners = np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
    
plt.imshow(flat_chess)
```

```python
corners=cv2.goodFeaturesToTrack(gray_real,64,0.01,10)
corners = np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
    
plt.imshow(real_chess)
```
